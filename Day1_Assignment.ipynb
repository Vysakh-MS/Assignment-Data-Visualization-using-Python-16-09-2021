```python
import pandas as pd
import numpy as pd
import matplotlib.pyplot as plt
```

```python
x=[1,2,3]
y=[2,3,4]
plt.figure(figsize=(5,4),dpi=100)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y,label='2x', color='green', marker='o', linestyle='dashed', markersize=7 , markeredgecolor='b' , linewidth=3)
plt.title("My first line chart!")
plt.savefig('myfirstlinegraph.png',dpi=300)
plt.legend()
plt.show()
```
